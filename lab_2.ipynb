```python
import pandas as pd

df1 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'key': ['K1', 'K2', 'K4', 'K5'],
    'C': ['C1', 'C2', 'C4', 'C5'],
    'D': ['D1', 'D2', 'D4', 'D5']
})

result = pd.merge(df1, df2, on='key', how='inner')

result
```
